### 8.3 PreProcessing predictor variables

```python
# #Below, I am splitting the dataframe into one with all (true) numeric variables, 
# #and another dataframe holding the (ordinal) factors.
ordinalVarNames = ['CentralAir', 'LandSlope', 'Street', 'PavedDrive',
              'PoolQC', 'FireplaceQu', 'LotShape', 'GarageFinish', 
              'GarageQual', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 
              'BsmtFinType1', 'BsmtFinType2', 'MasVnrType', 'KitchenQual', 
              'Functional', 'ExterQual', 'ExterCond', 'HeatingQC', 'IsNew',
              'NeighRich', 'Remod', 'OverallQual', 'OverallCond']
numericVarNames = [x for x in numericVarNames if x not in ordinalVarNames]
DFnumeric = all_df[numericVarNames]
DFfactors = all_df[factorVarNames + ordinalVarNames]
print('There are', len(DFnumeric.columns), 'numeric variables, and', len(DFfactors.columns), 'factor variables')
```

#### 8.3.1 Skewness and normalizing of the numeric predictors

```python
# Skewness is a measure of the symmetry in a distribution.
# In order to fix the skewness, I am taking the log for all 
# numeric predictors with an absolute skew greater than 0.8
from scipy import stats
col_DFnum_preds = [x for x in DFnumeric.columns if x != 'SalePrice']
for x in col_DFnum_preds:
    if np.abs(DFnumeric[x].skew()) > 0.8:
        DFnumeric[x] = np.log(DFnumeric[x]+1)
```

```python
DFnumeric.isnull().sum().sum()
```

```python
# Normalizing the data
from sklearn import preprocessing
x = DFnumeric[col_DFnum_preds].values
center_scaler = preprocessing.StandardScaler()
x_scaled = center_scaler.fit_transform(x)
DFnumeric[col_DFnum_preds] = pd.DataFrame(x_scaled)
```

#### 8.3.2 One hot encoding the categorical variables

```python
# Conducting one-hot encoding to categorical variables except for oridinal variables
DFdummies = pd.get_dummies(DFfactors[factorVarNames])
```

#### 8.3.3 Removing levels with few or no observations in train or test

```python
# check if some values are absent in the test set
ZerocolTest = []
for col in DFdummies.columns:
    if DFdummies[all_df['SalePrice'].isnull()][col].sum() == 0:
        ZerocolTest.append(col)

DFdummies = DFdummies.drop(ZerocolTest, axis = 1)

# check if some values are absent in the train set
ZerocolTrain = []
for col in DFdummies.columns:
    if DFdummies[~all_df['SalePrice'].isnull()][col].sum() < 10:
        ZerocolTrain.append(col)

DFdummies = DFdummies.drop(ZerocolTrain, axis = 1)
```

```python
# concat all dataframes to creat the final dataframe
final_df = pd.concat([DFdummies, DFnumeric, DFfactors[ordinalVarNames]], axis = 1)
```

## 9 Modeling [20]

### 9.1 Lasso regression model

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
import random

random.seed(666)
lassoreg = Lasso(alpha = 1)

X = train1.drop(['SalePrice'], axis = 1)
y = train1['SalePrice']

# define the parameter values that should be searched
alpha_range = list(np.arange(0,stop=1,step=0.01))

# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(alpha = alpha_range)

# instantiate the grid
grid = GridSearchCV(lassoreg, param_grid, cv=10, scoring='neg_mean_squared_error', n_jobs= -1)

# Run the grid search on the data
grid.fit(X, y)
```

```python
print(grid.best_params_)
print(grid.best_score_)
print(grid.best_estimator_)
```

```python
# Make Prediction on test set
X_test = test1.drop(['SalePrice'], axis = 1)
lassoreg = grid.best_estimator_
lassoreg.fit(X,y)
pred_lasso = lassoreg.predict(X_test)
# Prediction values need to reverse the log to the real values
pred_lasso = [np.exp(x) for x in pred_lasso]
```

### 9.2 XGBoost model

```python
import xgboost as xgb
from sklearn.model_selection import ShuffleSplit

xgb_model = xgb.XGBRegressor()

cv_sets_xgb = ShuffleSplit(n_splits=5) # shuffling our data for cross-validation

parameters_xgb = {'max_depth': [2,3,4,5,6],'min_child_weight':[1, 2, 3, 4 ,5],\
                    'n_estimators':[1000],'gamma': [0], 'subsample':[1],\
                  'learning_rate':[0.1, 0.05, 0.01], 'colsample_bytree': [1]}

scorer_xgb = 'neg_mean_squared_error'

grid_obj_xgb = GridSearchCV(xgb_model,parameters_xgb,scoring = scorer_xgb,cv = cv_sets_xgb,\
                            verbose = 10, n_jobs = -1)

grid_fit_xgb = grid_obj_xgb.fit(X, y)

xgb_opt = grid_fit_xgb.best_estimator_
```

```python
xgb_opt
```

```python
print(grid_fit_xgb.best_params_)
print(grid_fit_xgb.best_score_)
```

```python
# Make Prediction on test set
xgb_opt.fit(X,y)
pred_XGboost = xgb_opt.predict(X_test)
# Prediction values need to reverse the log to the real values
pred_XGboost = [np.exp(x) for x in pred_XGboost]
```

```python
# Make the variable importance plot using XGboost
xgb.plot_importance(xgb_opt, max_num_features=20)
```
